Exploratory Data Analysis (EDA)

# Gender

```python
palette = {'M': 'blue', 'F': 'yellow'}
ax = sns.countplot(x = 'Gender' , data= df, hue='Gender', palette=palette)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
# palette = {'M': 'blue', 'F': 'yellow'}
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen, hue= 'Gender')
```

From above graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than men

# State

```python
df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by ='Orders', ascending=False).head(10)
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by ='Orders', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(16,8)})
ax= sns.barplot(x='State', y='Orders', data = sales_state, hue='State')

for bar in ax.containers:
    ax.bar_label(bar, fontweight='bold')
```

```python
df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False).head(10)
```

```python
# total Amounts/Sales from top 10 states

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(16,8)})
ax= sns.barplot(x='State', y='Amount', data = sales_state, hue='State')

ax.set_xlabel('State', fontsize=14, fontweight='bold')
ax.set_ylabel('Amount', fontsize=14, fontweight='bold')
```

* From the above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra, and Karnataka
* Respectively but total sales/amount is from UP, Maharashtra and Karnataka

# Marital Status

```python
ax = sns.countplot(x = 'Marital_Status' , data = df, hue= 'Marital_Status')
sns.set(rc ={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars, fontweight='bold')
```

```python
sales_Shadi = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(5,5)})
ax= sns.barplot(x='Marital_Status', y='Amount', data = sales_Shadi, hue='Gender')
```

*From the above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
sns.set(rc={'figure.figsize':(20,6)})
ax = sns.countplot(data= df, x = 'Occupation', hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars , fontweight='bold')
```

```python
sales_Occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(16,8)})
ax= sns.barplot(x='Occupation', y='Amount', data = sales_Occupation, hue='Occupation')

ax.set_xlabel('Occupation', fontsize=14, fontweight='bold')
ax.set_ylabel('Amount', fontsize=14, fontweight='bold')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare amd Aviation sector*

# Product Category

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data= df, x = 'Product_Category', hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars , fontweight='bold')
```

```python
sales_Product_Category = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(20,8)})
ax= sns.barplot(x='Product_Category', y='Amount', data = sales_Product_Category, hue='Product_Category')

ax.set_xlabel('Product_Category', fontsize=14, fontweight='bold')
ax.set_ylabel('Amount', fontsize=14, fontweight='bold')
```

*From the above graphs we can see that most of the sold products are Food, Clothing and Electronic categories*

```python
sales_Product_ID = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by ='Orders', ascending=False).head(10)

sns.set(rc ={'figure.figsize':(16,8)})
ax= sns.barplot(x='Product_ID', y='Orders', data = sales_Product_ID, hue='Product_ID')

ax.set_xlabel('Product_ID', fontsize=14, fontweight='bold')
ax.set_ylabel('Orders', fontsize=14, fontweight='bold')
```
